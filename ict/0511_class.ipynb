```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            result.append(float(row[-1]))
    
plt.plot(result, 'hotpink')
plt.show()
```

```python
# 내 생일 최고 기온 데이터 시각화하기
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '21':
            result.append(float(row[-1]))
    
plt.plot(result, 'orange')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
result = []
for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
    
plt.hist(result, bins=1000, color='r')
plt.show()
```

```python
# 8월 최고 기온 데이터 히스토그램으로 표현하기
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
arg = []
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            arg.append(float(row[-1]))
    
plt.hist(arg, bins=1000, color='r')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
jan = []
aug = []
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':        
            jan.append(float(row[-1]))
        elif month == '08':
            aug.append(float(row[-1]))
    
plt.boxplot([jan, aug])
plt.show()
f.close()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
month = [[], [], [], [], [], [], [], [], [], [], [], []]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1]) - 1].append(float(row[-1]))

plt.boxplot(month)
plt.show()
f.close()
```
