```python
# swap
def swap(a, b):
    return b, a

print(swap(10, 20)) #결과값은 튜플로 반환된다
```

```python
# 변경 가능 객체를 인수로 전달할 경우
def g(t):
    t[0] = 0 # 리스트
    
a = [1, 2, 3]
g(a)
print(a)
```

```python
# 스코핑 룰
x = 1
def func(a):
    return a + x
    
def func2(a):
    x = 2
    return a + x

print(func(10))
print(func2(10))
print(x)
```

```python
# 스코핑 룰 (global)
g = 1
def func3(a):
    global g
    g = 3
    return a+ g
print(func3(10))
print(g)
```

```python
# 함수의 인수
def sum(a, b):
    return a + b
def incr(a, step = 1):
    return a + step

print(sum(2, 3))
print(incr(10))
print(incr(10, 2))
```

```python
# 키워드 인수
def area(width, height):
    return width * height

print(area(10, 12))
print(area(height = 4, width = 3)) #요렇게 코딩하는게 좋음
```

```python
# 가변인수
def get_total(*args):
    sum  = 0
    for x in args:
        sum += x
    return sum
print(get_total(1, 3, 5, 7, 9))
```

```python
# 사전 키워드 전달
def f (a, b, *args, **kwd):
    print(a, b)
    print(args)
    print(kwd)
    
f(10, 20, 30, 40, depth = 10, demesion = 3)
```

```python
# 함수 객체를 인수로 전달 가능
states = ['Alabama', 'Georgia', 'south carolina', 'West virginia']
def clean_string(strings, *funcs):
    results = []
    for string in strings:
        for func in funcs:
            string = func(string)
        results.append(string)
    return results

states = clean_string(states, str.strip, str.title) #.strip 문자열 및 공백 제거 / .title 첫글자 대문자
print(states)
```

```python
# 익명 함수 (Lambda)
def square(x):
    return x * 2

for i in range(10):
    print('{0} : {1}'.format(i, square(i)), end = ' ')
else:
    print()
```

```python
# Same as above with Lambda
for i in range(10):
    print('{0} : {1}\n'.format(i, (lambda x: x * 2) (i)), end = ' ')
else:
    print()
```

```python
# Lambda를 이용한 정렬
strings = ['foo', 'card', 'bar', 'abab', 'aaaa']
strings.sort(key = lambda x: len(x))
print(strings)
```

```python
strings.sort(key = lambda x: strings.count(x))
print(strings)
```

```python
# try, except, else
try:
    f = open('notexists.txt', 'r')
except FileNotFoundError as e:
    print(e)
else:
    data = f.read()
    f.close()
```

```python
# try ~ finally
f = open('result.txt', 'r')
try:
    print(f.read())
finally:
    f.close() #오류 여부와 상과없이 파일 close작업을 수행한다.
```

# 파일 입출력

```python
f = open('result.txt', 'r')
while True:
    line = f.readline()
    if not line:
        break
    print(line)
f.close()
```

```python
# 바이너리 파일 다루기
f = open('cat.png', 'rb')
print(f.read())
f.close()
```

```python
# 그 외 파일 관련 함수
f = open('result.txt', 'r', encoding = 'utf-8')

text = f.readline()
print(text)

pos = f.tell() #현재 파일 포인터를 얻어 옴
print(pos)

f.seek(16)
text = f.read()
print(text)
```
