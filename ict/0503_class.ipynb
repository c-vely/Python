# Matplot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3], [4,5,6])
plt.show()
```

```python
plt.plot([1,2,3], [4,5,6], label = '1st Line')
plt.plot([1,2,3], [2,4,6], label = '2st Line')
plt.xlabel('X Numbers')
plt.ylabel('Y Numbers')
plt.title('My Plot of X and Y')
plt.legend()
plt.show()
```

```python
import numpy as np
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
means = [20, 34, 30, 35, 27]

x = np.arange(len(labels))  # the label locations
rec = plt.bar(x, means, 0.5) # 막대의 두께: 0.5
plt.xticks(x, labels)
plt.show()
```

# Numpy

```python
import numpy as np
a = np.array([1, 2, 3])
print(a)
```

```python
# more thn one dimensions
aa = np.array([[1, 2], [3, 4]])
print(aa)
```

```python
# minimun dimesions
aaa = np.array([1, 2, 3, 4, 5], ndmin = 2)
print(aaa)
```

```python
# dtype parameter
a = np.array([1, 2, 3], dtype = complex)
print(a)
```

```python
# now apply it to ndarray object
dt = np.dtype([('age', np.int8)])
a = np.array([(10,), (20,), (30,)], dtype = dt)
a
```
