```python
# 지역명을 입력받아 데이터를 시각화 해보라.
f = open('age.csv')
data = csv.reader(f)
result = []
district = input("원하는 동을 입력해 주세요 : ")

for row in data:
    if district in row[0]:
        for i in row[3:]:
            result.append(int(i))
print(result)
plt.style.use('ggplot')
plt.plot(result)
plt.show()
f.close()
```

```python
# 지역명을 입력받아 데이터를 시각화 해보라.
f = open('age.csv')
data = csv.reader(f)
result = []
district = input("원하는 동을 입력해 주세요 : ")

for row in data:
    if district in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',', '')))
print(result)
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(district + '지역의 인구구조')
plt.plot(result)
plt.show()
f.close()
```

```python
# 항아리 모양 그래프 그리기
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '신림동' in row[0]:
        for i in range(0, 101):
            m.append(int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()
```

```python
file = open('gender.csv')
data = csv.reader(file)
m = []
f = []

for row in data:
    if '신림동' in row[0]:
        for i in row[3:104]:
                     m.append(int(i))
        for i in row[106:]:
                     f.append(int(i))
                     
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
file.close()           
```

```python
# 수평막대그래프로 시각화하기 
file = open('gender.csv')
data = csv.reader(file)
m = []
f = []

for row in data:
    if '신림동' in row[0]:
        for i in row[3:104]:
                     m.append(-int(i))
        for i in row[106:]:
                     f.append(int(i))
                
plt.rc('font', family = 'Malgun Gothic')
plt.title('신림동 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남자')      
plt.barh(range(101), f, label = '여자')
plt.legend()
plt.show()
file.close()           
```

```python
# 제주도에는 여성의 비율이 더 높을까?
file = open('gender.csv')
data = csv.reader(file)
name = '제주'
m = []
f = []

for row in data:
    if name in row[0]:
        for i in row[3:104]:
                     m.append(-int(i))
        for i in row[106:]:
                     f.append(int(i))
        break
                
plt.rc('font', family = 'Malgun Gothic')
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남자')      
plt.barh(range(101), f, label = '여자')
plt.legend()
plt.show()
file.close() 
```
