```python
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False
```

```python
f = open('subwaytime_202304.csv', encoding='UTF-8')
data = csv.reader(f)
next(data) # 상위 헤더 삭제
next(data) # 상위 헤더 삭제
result = []

for row in data:
    for i in range(4, len(row)):
        if row[i] == '': 
            row[i] = 0
    row[4:] = map(int, row[4:])
    result.append(row[10])
    
plt.bar(range(len(result)), result)
plt.show()
```

```python
# 아침 7~9시 승차 인원 더하기
f = open('subwaytime_202304.csv', encoding='UTF-8')
data = csv.reader(f)
next(data) # 상위 헤더 삭제
next(data) # 상위 헤더 삭제
result = []

for row in data:
    for i in range(4, len(row)):
        if row[i] == '': 
            row[i] = 0
    row[4:] = map(int, row[4:])
    result.append(sum(row[10:15:2]))
    
result.sort() 
plt.style.use('ggplot')
plt.figure(figsize=(15, 10))
plt.bar(range(len(result)), result)
plt.show()
```

```python
# 아침 7~9시 최고 승차인원수와승차역을 표시하라
f = open('subwaytime_202304.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
next(data)

mx = 0
mx_station = ''

for row in data:
    for i in range(4, len(row)):
        if row[i] == '': 
            row[i] = 0
    row[4:] = map(int, row[4:])
    if(sum(row[10:15:2])) > mx:
        mx = sum(row[10:15:2])
        mx_station = row[3] + '(' + row[1] + ')'
        
print(mx_station, mx)
```

```python
# 아침 7~9시 최고 하차 인원수와하차역을 표시하라
f = open('subwaytime_202304.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
next(data)

mx = 0
mx_station = ''

for row in data:
    for i in range(4, len(row)):
        if row[i] == '': 
            row[i] = 0
    row[4:] = map(int, row[4:])
    a = sum(row[10:15:2])
    if a > mx:
        mx = a
        mx_station = row[3] + '(' + row[1] + ')'
        
print(mx_station, mx)
f.close()
```

```python
# 입력된 시간대 중 사람들이 가장 많이 타고 내리는 역 찾기
f = open('subwaytime_202304.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
next(data)

mx = 0
mx_station = ''

t = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요? : '))

for row in data:
    for i in range(4, len(row)):
        if row[i] == '': 
            row[i] = 0
    row[4:] = map(int, row[4:])
    a = row[4 + (t-4) * 2]
    if a > mx:
        mx = a
        mx_station = row[3] + '(' + row[1] + ')'    
            
print(mx_station, mx)
f.close()
```

```python
# 시간대 사람들이 가장 많이 타고 내리는 역 찾고 그래프 그리기
f = open('subwaytime_202304.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
next(data)

mx = [0] * 24
mx_station = [''] * 24

for row in data:
    for i in range(4, len(row)):
        if row[i] == '': 
            row[i] = 0
    row[4:] = map(int, row[4:])
    
    for j in range(24):
        a = row[j * 2 + 4]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3]
print(mx_station, mx)

plt.rc('font', family='Malgun Gothic') 
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation = 90)
plt.show()
f.close()
```

```python
# 지하철 시간대별 승하차 인원 추이
f = open('subwaytime_202304.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
next(data)

s_in = [0] * 24
s_out = [0] * 24

t = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요? : '))

for row in data:
    for i in range(4, len(row)):
        if row[i] == '': 
            row[i] = 0
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[4 + i * 2]
        s_out[i] += row[5 + i * 2]
        
print(s_in)
print(s_out)

plt.figure(dpi = 300)            
plt.rc('font', family='Malgun Gothic') 
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24), range(4, 28))
plt.show()
f.close()
```
