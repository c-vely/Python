### 2.

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
setosa = x[y==0]
versicolor = x[y==1]
verginica = x[y==2]
```

```python
setosa[0:5,:]
```

### 3.

```python
import matplotlib.pyplot as plt
plt.scatter(setosa[:,0], setosa[:,1], label='Setosa')
plt.scatter(versicolor[:,0], versicolor[:,1], label='Versicolor')
plt.scatter(verginica[:,0], verginica[:,1], label='Verginica')
plt.legend(fontsize=10)
plt.show()
```

### 4.

```python
plt.scatter(setosa[:,2], setosa[:,3], label='Setosa')
plt.scatter(versicolor[:,2], versicolor[:,3], label='Versicolor')
plt.scatter(verginica[:,2], verginica[:,3], label='Verginica')
plt.legend(fontsize=10)
plt.show()
```
