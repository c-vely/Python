2번 문제

```python
import pandas as pd

df1 = pd.DataFrame([["cherry", "Fruit", 100],
                    ["mango", "Fruit", 110],
                    ["cherry", "Vegetable", 60],
                    ["cherry", "Vegetable", 80]],
                    columns = ["Name", "Type", "Price"]
                    )
df2 = pd.DataFrame([["pepper", "Vegetable", 50],
                    ["carrot", "Vegetable", 70],
                    ["banana", "Fruit", 90],
                    ["kiwi", "Fruit", 120]],
                    columns = ["Name", "Type", "Price"]
                    )

df3 = pd.concat([df1, df2], axis = 0)

print(df3)
print()
df_veg = df3[df3["Type"] == "Vegetable"]
df_veg = df_veg.sort_values(by = "Price", ascending = False) #가격으로 내림차순 정리

print(df_veg)
print()
df_Fruit = df3[df3["Type"] == "Fruit"]
df_Fruit = df_veg.sort_values(by = "Price", ascending = False) #가격으로 내림차순 정리

print(df_Fruit)
print()

print(df_veg.iloc[0]["Price"] + df_Fruit.iloc[0]["Price"])#iloc를 통해 첫번째 값 가져 옴
```

3번 문제

```python
import pandas as pd

idx = ["Sue", "Ryan", "Jay", "Jane", "Anna"]
col = ["round_1", "round_2", "round_3", "round_4", "round_5"]

data = [[55, 65, 60, 66, 57],
        [62, 77, 71, 79, 67],
        [88, 81, 79, 89, 77],
        [45, 35, 30, 46, 47],
        [91, 96, 90, 97, 99]]
df = pd.DataFrame(index = idx, columns = col, data = data)

col_round_6 = [11,15,13,17,19]
df["round_6"] = col_round_6

print(df)
print(df.describe().loc[["mean", "max", "min"]])
```

4번 문제

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)
red = [0 for values in t ]
green = [(1.7)**values for values in t] 
blue = [(2.66)**values for values in t]

plt.plot(t, red, c = "r", linestyle = "dashed")
plt.scatter(t, green, c = "g", marker = "s")
plt.scatter(t, blue, c = "b", marker = "^")
plt.show()
```

5번 문제

```python
import matplotlib.pyplot as plt

names = ["group_a","group_b","group_c"]
values = [1,10,100]

fig = plt.figure(figsize = (9,3))

ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

ax1.bar(names, values)

ax2.scatter(names, values)

ax3.plot(names, values)

plt.show()
```
