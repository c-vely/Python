# Q2

```python
import pandas as pd

df1 = pd.DataFrame([
    ["cherry", "Fruit", 100],
    ["mango", "Fruit", 110],
    ["potato", "Vegetable", 60],
    ["onion", "Vegetable", 80]],
    columns=["Name", "Type", "Price"])

df2 = pd.DataFrame([
    ["pepper", "Vegetable", 50],
    ["carrot", "Vegetable", 70],
    ["banana", "Fruit", 90],
    ["kiwi", "Fruit", 120]],
    columns=["Name", "Type", "Price"])

df3 = pd.concat([df1, df2], axis=0)

df_fruit = df3.loc[df3["Type"] == "Fruit"]
df_fruit = df_fruit.sort_values(by="Price", ascending=False)
print(df_fruit)

df_veg = df3.loc[df3["Type"] == "Vegetable"]
df_veg = df_veg.sort_values(by="Price", ascending=False)
print(df_veg)

print("Sum of Top 2 Fruit Price :", sum(df_fruit[:2]["Price"]))
print("Sum of Top 2 Vegetable Price :", sum(df_veg[:2]["Price"]))
```

```python
import matplotlib.pyplot as plt

names = ['group_a', 'group_b', 'group_c'] 
values = [1, 10, 100]

fig = plt.figure(figsize=(9, 3))
ax_1 = fig.add_subplot(1,3,1)
ax_2 = fig.add_subplot(1,3,2)
ax_3 = fig.add_subplot(1,3,3)

ax_1.bar(names, values)
ax_2.scatter(names, values)
ax_3.plot(names, values)
plt.show()
```
